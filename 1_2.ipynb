## Basic Operations on Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('demo_image.jpg')
```

```python
# accessing image shape
print(image.shape)
```

```python
# total num of pixel 
print(image.size)
```

```python
cv2.imshow("image",image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

## Splitting and Merging

```python
b,g,r=cv2.split(image) # cv2.split() is a costly operation , use it if only it is necessary
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.imshow(b)
plt.subplot(132)
plt.imshow(g)
plt.subplot(133)
plt.imshow(r)
```

```python
# matplotlib expects rgb format not bgr
# merging
image=cv2.merge((r,g,b)) 

plt.imshow(image)
```

```python
logo=cv2.imread("logo_demo1.jpg")

logo_gray=cv2.cvtColor(logo,cv2.COLOR_BGR2GRAY)
plt.imshow(logo_gray,cmap="gray")
```

```python
ret,thres_img=cv2.threshold(logo_gray,127,255,cv2.THRESH_BINARY) # thres_img gray suan 
plt.imshow(thres_img,cmap="gray") 
```

```python
black_logo=cv2.bitwise_not(thres_img)
plt.imshow(black_logo,cmap="gray") 
```
