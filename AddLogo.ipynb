## Necessary Libraries

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

## Read Logo

```python
logo_image = cv2.imread("resized_logo.jpg")
logo_rgb= cv2.cvtColor(logo_image,cv2.COLOR_BGR2RGB)
logo_gray= cv2.cvtColor(logo_image,cv2.COLOR_BGR2GRAY)
```

```python
logo_image.shape
```

```python
plt.figure(figsize=(12, 8)) 

plt.subplot(121)
plt.title("RGB image")
plt.imshow(logo_rgb)

plt.subplot(122)
plt.title("Gray scale image")
plt.imshow(logo_gray,cmap="gray")
```

## Create binary Image with threshold

```python
ret , binary_logo=cv2.threshold(logo_gray,110,255,cv2.THRESH_BINARY)

binary_logo_not=cv2.bitwise_not(binary_logo)

plt.figure(figsize=(12, 8)) 

plt.subplot(121)
plt.title("Binary logo")
plt.imshow(binary_logo,cmap="gray")

plt.subplot(122)
plt.title("Binary not logo")
plt.imshow(binary_logo_not,cmap="gray")

```

## Main Image

```python
main_image = cv2.imread("gs.jpg")
main_image_rgb= cv2.cvtColor(main_image,cv2.COLOR_BGR2RGB)
main_image_gray= cv2.cvtColor(main_image,cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(12, 8)) 

plt.subplot(121)
plt.title("RGB image")
plt.imshow(main_image_rgb)

plt.subplot(122)
plt.title("Gray scale image")
plt.imshow(main_image_gray,cmap="gray")
```

## Combine Images

```python
logo_width=logo_image.shape[0]
logo_height=logo_image.shape[1]
```

```python
# convert it to rgb because it is gonna be combined with rgb image
logo_not_rgb=cv2.cvtColor(logo_not,cv2.COLOR_GRAY2RGB)
plt.imshow(logo_not_rgb)
```

```python
roi=main_image_rgb[:logo_width,:logo_height]
plt.imshow(roi)
```

```python
logo_not_main_image=cv2.bitwise_and(logo_not_rgb,roi)
plt.imshow(logo_not_main_image)
```

```python
logo_and_roi=(cv2.bitwise_or(logo_rgb,logo_not_main_image))
plt.imshow(logo_and_roi)
```

```python
main_image_rgb[:logo_width,:logo_height,:]=logo_and_roi
```
